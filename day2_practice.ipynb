b. Slicing:
Syntax:
    variable(Starting index : Ending index)

```python
b = "Paris" # The end part is exclusive in slicing
b[1:4]
```

```python
b[::-1] # This will reverse the given data temporarily
```

a. List : List is mutable data structure so, we can access and change the data if required.

```python
lst = [1, 2, 3, 4, True, False, "Banana", 70]
lst[3]
```

```python
lst.append(99) # list.append(value or elements) this helps to add any value or elements at the end of the list 
```

```python
lst
```

```python
lst.extend([12, 13, 14]) # This will add only a set of value or data but it will not store a single value in a set
lst
```

```python
lst.reverse() # This will reverse your list permanently
lst
```

```python
lst.pop() # This will remove the last data or value from your list
lst
```

```python
lst.remove("Banana") # this will helps to remove the specific value or data which you have given 
lst
```

```python
lst.insert(3, "Apple") # This will helps you to insert specific data which you have given in the list
lst
```

```python
lst[0] = "America" # This will helps you to replace the specific data from your list to the given or inputed once
lst
```
